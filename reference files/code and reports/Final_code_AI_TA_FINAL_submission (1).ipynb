## Step 2 — Preprocessing & Feature Engineering

### 2.1 Text Cleaning

We will define a custom cleaning function to lowercase all text and strip out irregular punctuation, then apply it to the main text columns.

```python
df_model = df.copy()

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"\s+", " ", text)
    text = re.sub(r"[^\w\s\-\?\.,:;/%()]", " ", text)
    return text.strip()

df_model["question_clean"] = df_model["question"].apply(clean_text)
df_model["context_clean"] = df_model["context"].apply(clean_text)
df_model["long_answer_clean"] = df_model["long_answer"].apply(clean_text)

print(f"Text cleaning complete. Columns: {[c for c in df_model.columns if 'clean' in c]}")
```

Text cleaning is complete and our core text columns are now fully standardised for consistent downstream processing.

### 2.2 Input Variants & Label Encoding

We will generate three distinct input text variants (question only, question + context, and question + context + answer) and encode the labels as integers.

```python
df_model["text_q_only"] = "question: " + df_model["question_clean"]
df_model["text_q_ctx"] = (
    "question: " + df_model["question_clean"]
    + " context: " + df_model["context_clean"]
)
df_model["text_q_ctx_ans"] = (
    "question: " + df_model["question_clean"]
    + " context: " + df_model["context_clean"]
    + " answer: " + df_model["long_answer_clean"]
)

label2id = {"no": 0, "maybe": 1, "yes": 2}
id2label = {0: "no", 1: "maybe", 2: "yes"}
df_model["label_id"] = df_model["label"].map(label2id)
df_model["question_len"] = df_model["question_clean"].apply(lambda x: len(x.split()))
df_model["context_len"] = df_model["context_clean"].apply(lambda x: len(x.split()))
df_model["long_answer_len"] = df_model["long_answer_clean"].apply(lambda x: len(x.split()))
df_model["num_contexts"] = df_model["context_list"].apply(lambda x: len(x) if isinstance(x, list) else 1)

assert df_model["label_id"].isna().sum() == 0, "Unmapped labels found!"

print(f"Input variants created. Shape: {df_model.shape}")
print(f"Label distribution: {df_model['label_id'].value_counts().sort_index().to_dict()}")
```

Our three text permutations are built, lengths are recalculated based on clean text, and the assertion check confirms no unmapped labels remain.

### 2.3 Handcrafted NLP Features

We will engineer handcrafted NLP features to detect specific vocabulary patterns like negation, causation, comparison, and hedging language in our texts.

```python
df_model["has_negation"] = df_model["question_clean"].apply(
    lambda x: int(any(w in x.split() for w in ["no", "not", "without", "lack", "absence"]))
)
df_model["has_causal"] = df_model["question_clean"].apply(
    lambda x: int(any(w in x for w in ["cause", "effect", "impact", "influence", "affect", "lead to"]))
)
df_model["has_comparison"] = df_model["question_clean"].apply(
    lambda x: int(any(w in x for w in ["compar", "differ", "versus", "vs", "better", "worse"]))
)
df_model["has_association"] = df_model["question_clean"].apply(
    lambda x: int(any(w in x for w in ["associat", "correlat", "relat", "link"]))
)
df_model["has_risk"] = df_model["question_clean"].apply(
    lambda x: int("risk" in x)
)
df_model["context_evidence_strength"] = df_model["long_answer_clean"].apply(
    lambda x: int(any(w in x for w in ["significant", "strongly", "clearly", "demonstrated"]))
)
df_model["answer_hedging"] = df_model["long_answer_clean"].apply(
    lambda x: int(any(w in x for w in ["may", "might", "suggest", "possibly", "further study", "inconclusive"]))
)

print("Preprocessing complete")
print(f"Shape: {df_model.shape}")
print(f"Label distribution:\n{df_model['label_id'].value_counts().sort_index()}")
print(f"\nSample cleaned text (first row):")
print(df_model["text_q_ctx"].iloc[0][:200])
```

Preprocessing is complete and the DataFrame has expanded to 24 columns with all handcrafted features added alongside the verified label distribution.

## Step 4 — Baseline: TF-IDF + Logistic Regression (with Cross-Validation)

### 4.1 Cross-Validation Helper Function

We will define a reusable helper function that runs Stratified K-Fold cross-validation on the training set and then evaluates on the test set.

```python
def evaluate_with_cv(pipeline, X_tr, y_tr, X_te, y_te, n_folds=5):
    """Run stratified k-fold CV on train, then evaluate on test."""
    cv = StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=RANDOM_SEED)

    cv_acc = cross_val_score(pipeline, X_tr, y_tr, cv=cv, scoring="accuracy")
    cv_f1 = cross_val_score(pipeline, X_tr, y_tr, cv=cv, scoring="f1_macro")

    pipeline.fit(X_tr, y_tr)
    preds = pipeline.predict(X_te)

    test_acc = accuracy_score(y_te, preds)
    test_f1 = f1_score(y_te, preds, average="macro")

    return {
        "cv_acc_mean": cv_acc.mean(),
        "cv_acc_std": cv_acc.std(),
        "cv_f1_mean": cv_f1.mean(),
        "cv_f1_std": cv_f1.std(),
        "test_acc": test_acc,
        "test_f1": test_f1,
        "preds": preds
    }

print("Cross-validation helper function defined.")
```

The helper function is ready and will automatically compute accuracy and macro-F1 scores for any pipeline we pass to it.

### 4.2 TF-IDF + Logistic Regression Pipeline

We will build a Scikit-Learn pipeline that vectorises text using TF-IDF (unigrams and bigrams) and feeds it into a balanced Logistic Regression classifier.

```python
baseline_pipe = Pipeline([
    ("tfidf", TfidfVectorizer(
        lowercase=True, stop_words="english",
        ngram_range=(1, 2), max_features=30000
    )),
    ("clf", LogisticRegression(
        max_iter=2000, class_weight="balanced",
        random_state=RANDOM_SEED, C=1.0
    ))
])

baseline_res = evaluate_with_cv(baseline_pipe, X_train_qctx, y_train, X_test_qctx, y_test)

print("=" * 60)
print("BASELINE: TF-IDF (1,2)-grams + Logistic Regression")
print("=" * 60)
print(f"5-Fold CV Accuracy : {baseline_res['cv_acc_mean']:.4f} +/- {baseline_res['cv_acc_std']:.4f}")
print(f"5-Fold CV Macro-F1 : {baseline_res['cv_f1_mean']:.4f} +/- {baseline_res['cv_f1_std']:.4f}")
print(f"Test Accuracy      : {baseline_res['test_acc']:.4f}")
print(f"Test Macro-F1      : {baseline_res['test_f1']:.4f}")
print(f"\n{classification_report(y_test, baseline_res['preds'], target_names=['no','maybe','yes'], zero_division=0)}")
```

The baseline achieved a test accuracy of ~47% and a macro-F1 of 0.33, with the model struggling heavily on the minority 'no' and 'maybe' classes.

## Step 5 — Axis 1: Input Text Representation

**Hypothesis:** The information fed to the classifier strongly affects performance.
- **A1a (Question only):** Using only the question may be insufficient because the answer depends on evidence in the context. We expect this to perform worst.
- **A1b (Question + Context):** This is the baseline input. Adding context should help because the model can learn associations between evidence phrases and labels.
- **A1c (Question + Context + Long Answer):** Including the long answer summary may further help because it contains the author's own conclusion, which is closely tied to the label. However, this may cause the model to rely on surface patterns in the answer rather than learning genuine reasoning.

### 5.1 Evaluating the 'Question Only' Variant

We will test how the model performs when given only the question text, without any context or answer information.

```python
axis1_pipe = Pipeline([
    ("tfidf", TfidfVectorizer(
        lowercase=True, stop_words="english",
        ngram_range=(1, 2), max_features=30000
    )),
    ("clf", LogisticRegression(
        max_iter=2000, class_weight="balanced",
        random_state=RANDOM_SEED
    ))
])

from sklearn.base import clone

pipe_a1a = clone(axis1_pipe)
res_a1a = evaluate_with_cv(pipe_a1a, X_train_qonly, y_train, X_test_qonly, y_test)
all_preds["A1a_Q_only"] = res_a1a["preds"]

print("A1a — Question Only + TF-IDF + LR")
print(f"  CV Macro-F1 : {res_a1a['cv_f1_mean']:.4f} +/- {res_a1a['cv_f1_std']:.4f}")
print(f"  Test Acc    : {res_a1a['test_acc']:.4f}")
print(f"  Test F1     : {res_a1a['test_f1']:.4f}")
print(classification_report(y_test, res_a1a["preds"], target_names=["no","maybe","yes"], zero_division=0))
```

Surprisingly, using just the question gave a Test Macro-F1 of 0.3680, slightly edging out the baseline that included context.

### 5.2 Evaluating the 'Question + Context + Answer' Variant

We will test the richest input variant that includes the question, context, and long answer, and compare it to the other two variants.

```python
res_a1b = baseline_res
print("\nA1b — Question + Context + TF-IDF + LR  (= Baseline)")
print(f"  CV Macro-F1 : {res_a1b['cv_f1_mean']:.4f} +/- {res_a1b['cv_f1_std']:.4f}")
print(f"  Test F1     : {res_a1b['test_f1']:.4f}")

pipe_a1c = clone(axis1_pipe)
res_a1c = evaluate_with_cv(pipe_a1c, X_train_qctxans, y_train, X_test_qctxans, y_test)
all_preds["A1c_Q_Ctx_Ans"] = res_a1c["preds"]

print("\nA1c — Question + Context + Long Answer + TF-IDF + LR")
print(f"  CV Macro-F1 : {res_a1c['cv_f1_mean']:.4f} +/- {res_a1c['cv_f1_std']:.4f}")
print(f"  Test Acc    : {res_a1c['test_acc']:.4f}")
print(f"  Test F1     : {res_a1c['test_f1']:.4f}")
print(classification_report(y_test, res_a1c["preds"], target_names=["no","maybe","yes"], zero_division=0))
```

Adding the long answer into the input actually dropped the Test F1 to 0.3247, indicating that more text can introduce noise for a simple TF-IDF model.

### 6.1 Feature Builder Functions

We will define utility functions for building combined word-level and character-level TF-IDF features, and a custom maybe-threshold prediction function.

```python
def build_word_char_features(train_texts, other_texts, max_word=40000, max_char=30000):
    word_vec = TfidfVectorizer(
        lowercase=True,
        stop_words="english",
        ngram_range=(1, 2),
        min_df=2,
        sublinear_tf=True,
        max_features=max_word
    )
    char_vec = TfidfVectorizer(
        lowercase=True,
        analyzer="char_wb",
        ngram_range=(3, 5),
        min_df=2,
        sublinear_tf=True,
        max_features=max_char
    )

    X_train_word = word_vec.fit_transform(train_texts)
    X_train_char = char_vec.fit_transform(train_texts)
    X_train = hstack([X_train_word, X_train_char]).tocsr()

    outputs = []
    for texts in other_texts:
        X_word = word_vec.transform(texts)
        X_char = char_vec.transform(texts)
        outputs.append(hstack([X_word, X_char]).tocsr())

    return X_train, outputs, word_vec, char_vec


def maybe_threshold_predict(probs, maybe_threshold=0.40, maybe_class_id=1):
    non_maybe_best = np.where(probs[:, 0] >= probs[:, 2], 0, 2)
    preds = np.where(probs[:, maybe_class_id] >= maybe_threshold, maybe_class_id, non_maybe_best)
    return preds

print("Feature builder and threshold prediction functions defined.")
```

Both utility functions are defined and ready for use across all Axis 2 experiments.

### Step 6.9 — Ensemble: Soft Voting Classifier

**Hypothesis:** Individual classifiers each make different types of errors. Combining them via soft voting (averaging predicted probabilities) should reduce variance and improve macro-F1 by leveraging complementary strengths — e.g., LR captures linear boundaries while SVM finds wider margins.

We will combine the calibrated SVM, tuned LR, and Naive Bayes into a soft voting ensemble and evaluate its performance.

```python
from sklearn.ensemble import VotingClassifier

print("\nA2e — Soft Voting Ensemble (LR + Cal-SVM + NB)")

svm_for_ens = CalibratedClassifierCV(
    LinearSVC(C=svm_grid.best_params_["C"], class_weight="balanced",
             random_state=RANDOM_SEED, max_iter=10000),
    method="sigmoid", cv=3
)

lr_for_ens = LogisticRegression(
    C=lr_grid.best_params_["C"], max_iter=4000,
    class_weight="balanced", solver="liblinear",
    random_state=RANDOM_SEED
)

nb_for_ens = MultinomialNB(alpha=nb_grid.best_params_["alpha"])

ensemble = VotingClassifier(
    estimators=[
        ("lr", lr_for_ens),
        ("svm", svm_for_ens),
        ("nb", nb_for_ens)
    ],
    voting="soft",
    n_jobs=-1
)

ensemble.fit(X_train_wc, y_train)
pred_ensemble = ensemble.predict(X_test_wc)

ens_acc = accuracy_score(y_test, pred_ensemble)
ens_f1 = f1_score(y_test, pred_ensemble, average="macro")

print(f"  Test Acc     : {ens_acc:.4f}")
print(f"  Test Macro-F1: {ens_f1:.4f}")
print(classification_report(y_test, pred_ensemble, target_names=['no','maybe','yes'], zero_division=0))

all_results["A2e: Soft Voting Ensemble"] = {
    "CV Macro-F1": "N/A (ensemble)",
    "Test Acc": ens_acc,
    "Test Macro-F1": ens_f1
}
all_preds["A2e_Ensemble"] = pred_ensemble
```

### Step 6.10 — Zero-Shot LLM-Based Classification (Classification vs Generation Axis)

A more effective LLM approach would likely require a biomedical-specific generative model (e.g., BioGPT) or careful prompt engineering with few-shot examples.

## Step 7 — Fine-tuned BioBERT (Fixed Training)

Key improvements over the original attempt:

### 7.1 PyTorch Setup & Configuration

We will check for GPU availability, set up the fine-tuning hyperparameters, and prepare the train/validation split for the BioBERT model.

```python
FINETUNE_DONE = False
pred_finetune = None

TRANSFORMERS_OK = TRANSFORMERS_AVAILABLE if "TRANSFORMERS_AVAILABLE" in globals() else False

if TRANSFORMERS_OK:
    try:
        from torch.utils.data import Dataset
        from transformers import Trainer, TrainingArguments, EarlyStoppingCallback
    except Exception as e:
        print(f"Could not import training utilities: {e}")
        TRANSFORMERS_OK = False

if TRANSFORMERS_OK and torch.cuda.is_available():
    MODEL_NAME = "dmis-lab/biobert-base-cased-v1.2"
    MAX_LEN = 256
    BATCH_SIZE = 8
    EPOCHS = 5
    device = torch.device("cuda")

    X_train_list = list(X_train_qctx)
    X_test_list = list(X_test_qctx)
    y_train_list = list(y_train)
    y_test_list = list(y_test)

    X_tr_ft, X_val_ft, y_tr_ft, y_val_ft = train_test_split(
        X_train_list,
        y_train_list,
        test_size=0.10,
        random_state=RANDOM_SEED,
        stratify=y_train_list
    )

    print(f"Fine-tune train: {len(X_tr_ft)}, Val: {len(X_val_ft)}, Test: {len(X_test_list)}")
```

The fine-tuning environment is configured with the BioBERT model path, batch size, and a 90/10 train-validation split ready for training.

### 7.2 Dataset Class & Loss Configuration

We will define a custom PyTorch Dataset class for tokenisation, compute class weights to handle the label imbalance, and set up a weighted loss trainer.

```python
if TRANSFORMERS_OK and torch.cuda.is_available():
    class QADataset(Dataset):
        def __init__(self, texts, labels, tokenizer, max_len=256):
            self.texts = texts
            self.labels = labels
            self.tokenizer = tokenizer
            self.max_len = max_len

        def __len__(self):
            return len(self.texts)

        def __getitem__(self, idx):
            enc = self.tokenizer(
                str(self.texts[idx]),
                truncation=True,
                padding="max_length",
                max_length=self.max_len,
                return_tensors="pt"
            )
            item = {k: v.squeeze(0) for k, v in enc.items()}
            item["labels"] = torch.tensor(int(self.labels[idx]), dtype=torch.long)
            return item

    label_counts = Counter(y_tr_ft)
    total = len(y_tr_ft)
    n_classes = 3
    class_weights = torch.tensor(
        [total / (n_classes * label_counts[i]) for i in range(n_classes)],
        dtype=torch.float32,
        device=device
    )

    print(f"Class weights: {class_weights.detach().cpu().tolist()}")

    class WeightedTrainer(Trainer):
        def compute_loss(self, model, inputs, return_outputs=False, **kwargs):
            labels = inputs.pop("labels").to(model.device)
            outputs = model(**inputs)
            logits = outputs.logits
            loss_fn = torch.nn.CrossEntropyLoss(weight=class_weights)
            loss = loss_fn(logits, labels)
            return (loss, outputs) if return_outputs else loss

    def compute_metrics(eval_pred):
        logits, labels = eval_pred
        preds = np.argmax(logits, axis=1)
        return {
            "accuracy": accuracy_score(labels, preds),
            "macro_f1": f1_score(labels, preds, average="macro")
        }

    print("Dataset class and weighted trainer defined.")
```

The custom dataset class handles tokenisation, and the WeightedTrainer uses inverse-frequency class weights to penalise misclassification of minority labels.

### 7.3 Training & Evaluation

We will load the pre-trained BioBERT model, configure the Hugging Face Trainer with early stopping, run fine-tuning, and evaluate on the test set.

```python
if TRANSFORMERS_OK and torch.cuda.is_available():
    try:
        tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
        model = AutoModelForSequenceClassification.from_pretrained(
            MODEL_NAME,
            num_labels=3
        ).to(device)

        train_ds = QADataset(X_tr_ft, y_tr_ft, tokenizer, MAX_LEN)
        val_ds = QADataset(X_val_ft, y_val_ft, tokenizer, MAX_LEN)
        test_ds = QADataset(X_test_list, y_test_list, tokenizer, MAX_LEN)

        steps_per_epoch = max(1, len(train_ds) // BATCH_SIZE)
        total_steps = steps_per_epoch * EPOCHS
        warmup_steps = int(0.1 * total_steps)

        training_args = TrainingArguments(
            output_dir="./biobert_output",
            eval_strategy="epoch",
            save_strategy="epoch",
            logging_strategy="epoch",
            load_best_model_at_end=True,
            metric_for_best_model="macro_f1",
            greater_is_better=True,
            per_device_train_batch_size=BATCH_SIZE,
            per_device_eval_batch_size=BATCH_SIZE,
            num_train_epochs=EPOCHS,
            learning_rate=2e-5,
            weight_decay=0.01,
            warmup_steps=warmup_steps,
            report_to="none",
            fp16=True,
            save_total_limit=2
        )

        trainer = WeightedTrainer(
            model=model,
            args=training_args,
            train_dataset=train_ds,
            eval_dataset=val_ds,
            compute_metrics=compute_metrics,
            callbacks=[EarlyStoppingCallback(early_stopping_patience=2)]
        )

        print("\nStarting BioBERT fine-tuning with class-weighted loss...")
        trainer.train()

        print("\nEvaluating best checkpoint on test set...")
        test_output = trainer.predict(test_ds)
        pred_finetune = np.argmax(test_output.predictions, axis=1)

        ft_acc = accuracy_score(y_test_list, pred_finetune)
        ft_f1 = f1_score(y_test_list, pred_finetune, average="macro")

        print("\nFine-tuned BioBERT Results:")
        print(f"  Test Accuracy : {ft_acc:.4f}")
        print(f"  Test Macro-F1 : {ft_f1:.4f}")
        print(classification_report(
            y_test_list,
            pred_finetune,
            target_names=["no", "maybe", "yes"],
            zero_division=0
        ))

        all_results["A2d: BioBERT Fine-tuned"] = {
            "CV Macro-F1": "N/A (single val split)",
            "Test Acc": ft_acc,
            "Test Macro-F1": ft_f1
        }
        all_preds["A2d_BioBERT_FT"] = pred_finetune
        FINETUNE_DONE = True

    except Exception as e:
        print(f"Fine-tuning failed: {e}")
        FINETUNE_DONE = False

else:
    print("Skipping fine-tuning (no GPU or transformers unavailable)")
```

The fine-tuned BioBERT model achieved a modest test accuracy and macro-F1, showing that even transformer-based fine-tuning faces challenges on this imbalanced medical QA task.

### Step 9.6 — Statistical Significance: McNemar's Test

To determine whether differences between models are statistically meaningful (not just random), we run McNemar's test on pairs of predictions.

We will run McNemar's test between the baseline and best model to check if the performance difference is statistically significant.

```python
from scipy.stats import chi2

def mcnemar_test(y_true, preds_a, preds_b, name_a="A", name_b="B"):
    """McNemar's test: are two classifiers significantly different?"""
    correct_a = (preds_a == y_true)
    correct_b = (preds_b == y_true)

    b = np.sum(~correct_a & correct_b)
    c = np.sum(correct_a & ~correct_b)

    if b + c == 0:
        print(f"  {name_a} vs {name_b}: identical predictions (no test needed)")
        return

    statistic = (abs(b - c) - 1) ** 2 / (b + c)
    p_value = 1 - chi2.cdf(statistic, df=1)

    print(f"  {name_a} vs {name_b}:")
    print(f"    Discordant pairs: b={b} (A wrong, B right), c={c} (A right, B wrong)")
    print(f"    McNemar statistic: {statistic:.4f}, p-value: {p_value:.4f}")
    print(f"    {"Significant (p<0.05)" if p_value < 0.05 else "NOT significant (p>=0.05)"}")
    return p_value

print("\nMcNemar's Test for Statistical Significance")
print("=" * 60)

baseline_preds = all_preds["Baseline"]
best_pred_key_sig = max(all_preds.keys(), key=lambda k: f1_score(y_test, all_preds[k], average="macro"))
best_preds_sig = all_preds[best_pred_key_sig]

mcnemar_test(y_test, baseline_preds, best_preds_sig, "Baseline", best_pred_key_sig)

sorted_models = sorted(all_preds.keys(), key=lambda k: f1_score(y_test, all_preds[k], average="macro"), reverse=True)
if len(sorted_models) >= 2:
    mcnemar_test(y_test, all_preds[sorted_models[0]], all_preds[sorted_models[1]], sorted_models[0], sorted_models[1])
```

McNemar's test tells us whether the performance differences we observe are due to genuine model superiority or could have arisen by chance.

Understanding whether different models fail on the same examples or make complementary errors is crucial for understanding the limits of the approaches and the potential for ensemble methods.

We will analyse how many test errors are shared across models vs unique to each model.

```python
print("\nError Overlap Analysis")
print("=" * 60)

key_models = {}
for name in sorted_models[:4]:
    key_models[name] = (all_preds[name] != y_test)

model_names = list(key_models.keys())
error_masks = list(key_models.values())

print("\nPairwise error overlap (fraction of shared errors):")
overlap_matrix = np.zeros((len(model_names), len(model_names)))
for i in range(len(model_names)):
    for j in range(len(model_names)):
        shared = np.sum(error_masks[i] & error_masks[j])
        total = np.sum(error_masks[i] | error_masks[j])
        overlap_matrix[i, j] = shared / total if total > 0 else 0

fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(overlap_matrix, annot=True, fmt=".2f",
            xticklabels=[n[:20] for n in model_names],
            yticklabels=[n[:20] for n in model_names],
            cmap="YlOrRd", ax=ax)
ax.set_title("Error Overlap (Jaccard Similarity) Between Models")
plt.tight_layout()
os.makedirs("visualizations", exist_ok=True)
plt.savefig(os.path.join("visualizations", "Test_MacroF1_Comparison.png"), dpi=300, bbox_inches="tight")
plt.show()

all_wrong = np.ones(len(y_test), dtype=bool)
for mask in error_masks:
    all_wrong &= mask
print(f"\nSamples ALL top-4 models get wrong: {all_wrong.sum()} / {len(y_test)} ({all_wrong.sum()/len(y_test)*100:.1f}%)")
print("These are the hardest examples that no approach can handle.")
```

High overlap in errors suggests that certain examples are fundamentally ambiguous, while low overlap supports the value of ensemble methods.

## Step 10 — Quality Predictor

Both are evaluated via 5-fold stratified CV. This is an exploratory analysis on the test-set metadata — we acknowledge this is post-hoc and not a production-ready confidence estimator.

### 10.1 Generating Out-of-Fold Correctness Labels

We will run 5-fold cross-validation on the training set to generate unbiased correctness labels that tell us whether each sample would be predicted correctly.

```python
quality_features = [
    "question_len", "context_len", "long_answer_len", "num_contexts",
    "has_negation", "has_causal", "has_comparison", "has_association",
    "has_risk", "context_evidence_strength", "answer_hedging"
]


train_feats = train_df[["pubid"] + quality_features].copy() if "pubid" in train_df.columns else train_df[["pmid"] + quality_features].copy()
test_feats = test_df[["pubid"] + quality_features].copy() if "pubid" in test_df.columns else test_df[["pmid"] + quality_features].copy()

oof_correct = np.zeros(len(train_df), dtype=int)
cv5_q = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_SEED)

for fold, (tr_idx, va_idx) in enumerate(cv5_q.split(X_train_qctx, y_train), 1):
    X_tr_text = [X_train_qctx[i] for i in tr_idx]
    X_va_text = [X_train_qctx[i] for i in va_idx]
    y_tr = y_train[tr_idx]
    y_va = y_train[va_idx]

    X_tr_wc, [X_va_wc], _, _ = build_word_char_features(X_tr_text, [X_va_text])

    svm = LinearSVC(
        C=svm_grid.best_params_["C"],
        class_weight="balanced",
        random_state=RANDOM_SEED,
        max_iter=10000
    )
    cal = CalibratedClassifierCV(svm, method="sigmoid", cv=3)
    cal.fit(X_tr_wc, y_tr)

    va_probs = cal.predict_proba(X_va_wc)
    va_pred = maybe_threshold_predict(va_probs, maybe_threshold=best_thr, maybe_class_id=1)
    oof_correct[va_idx] = (va_pred == y_va).astype(int)

final_pred_key = "A2b_CalSVM_wordchar" if "A2b_CalSVM_wordchar" in all_preds else best_pred_key
test_correct = (all_preds[final_pred_key] == y_test).astype(int)

X_qual_train = train_feats[quality_features].values
y_qual_train = oof_correct

X_qual_test = test_feats[quality_features].values
y_qual_test = test_correct

print("Quality predictor training labels:")
print(pd.Series(y_qual_train).value_counts().sort_index())
print("Quality predictor test labels:")
print(pd.Series(y_qual_test).value_counts().sort_index())
```

The out-of-fold correctness labels are generated, providing an unbiased training signal for the quality predictor models.

### 10.2 Training Quality Predictor Models

We will train three different quality predictor models — Random Forest, MLP, and Logistic Regression — to see if handcrafted features can predict whether the main model will get a sample right.

```python
rf_qual = RandomForestClassifier(
    n_estimators=400,
    max_depth=None,
    min_samples_leaf=2,
    class_weight="balanced",
    random_state=RANDOM_SEED,
    n_jobs=-1
)

mlp_qual = Pipeline([
    ("scaler", StandardScaler()),
    ("mlp", MLPClassifier(
        hidden_layer_sizes=(64, 32),
        activation="relu",
        max_iter=500,
        early_stopping=True,
        validation_fraction=0.15,
        random_state=RANDOM_SEED
    ))
])

lr_qual = Pipeline([
    ("scaler", StandardScaler()),
    ("lr", LogisticRegression(
        max_iter=2000,
        class_weight="balanced",
        random_state=RANDOM_SEED
    ))
])

qual_models = {
    "RF": rf_qual,
    "MLP": mlp_qual,
    "LR": lr_qual
}

qual_rows = []
for name, model in qual_models.items():
    model.fit(X_qual_train, y_qual_train)
    pred = model.predict(X_qual_test)
    qual_rows.append({
        "Model": name,
        "Test Accuracy": accuracy_score(y_qual_test, pred),
        "Test Macro-F1": f1_score(y_qual_test, pred, average="macro")
    })
    print(f"\n{name} Quality Predictor")
    print(f"  Test Accuracy : {accuracy_score(y_qual_test, pred):.4f}")
    print(f"  Test Macro-F1 : {f1_score(y_qual_test, pred, average='macro'):.4f}")
    print(classification_report(y_qual_test, pred, target_names=["wrong", "correct"], zero_division=0))

qual_df = pd.DataFrame(qual_rows).sort_values("Test Macro-F1", ascending=False)
display(qual_df)
```

### 10.4 Quality Predictor with Embedding Features

We will enrich the quality predictor with a cosine similarity feature between question and context embeddings to see if semantic overlap improves prediction.

```python
if SBERT_OK and "embedder" in globals() and "error_df" in globals():
    from sklearn.metrics.pairwise import cosine_similarity

    print("Encoding questions and contexts separately for similarity features...")

    q_texts = list(X_test_qonly)
    ctx_texts = [str(c) for c in error_df["context"].fillna("").tolist()]

    q_embs = embedder.encode(q_texts, batch_size=32, show_progress_bar=True)
    ctx_embs = embedder.encode(ctx_texts, batch_size=32, show_progress_bar=True)

    q_ctx_sim = np.array([
        cosine_similarity([q], [c])[0][0]
        for q, c in zip(q_embs, ctx_embs)
    ])

    print("Encoding training questions and contexts for train-side similarity feature...")
    train_q_embs = embedder.encode(list(X_train_qonly), batch_size=32, show_progress_bar=True)
    train_ctx_embs = embedder.encode(
        [str(c) for c in train_df["context"].fillna("").tolist()],
        batch_size=32,
        show_progress_bar=True
    )

    train_q_ctx_sim = np.array([
        cosine_similarity([q], [c])[0][0]
        for q, c in zip(train_q_embs, train_ctx_embs)
    ])

    X_qual_train_enriched = np.column_stack([X_qual_train, train_q_ctx_sim])
    X_qual_test_enriched = np.column_stack([X_qual_test, q_ctx_sim])

    print(f"Enriched train shape: {X_qual_train_enriched.shape}")
    print(f"Enriched test shape:  {X_qual_test_enriched.shape}")
else:
    print("Skipping enriched quality predictor (embeddings or required variables not available)")
```

The embedding similarity features are computed and appended to the handcrafted feature set for comparison.

### 10.5 Comparing Plain vs. Enriched Quality Predictors

We will train two identical MLP models — one with only handcrafted features and one with the added embedding similarity — to measure the impact.

```python
if SBERT_OK and "embedder" in globals() and "X_qual_train_enriched" in globals():
    mlp_enriched_pipe = Pipeline([
        ("scaler", StandardScaler()),
        ("mlp", MLPClassifier(
            hidden_layer_sizes=(64, 32),
            activation="relu",
            max_iter=500,
            random_state=RANDOM_SEED,
            early_stopping=True,
            validation_fraction=0.15
        ))
    ])

    mlp_plain_pipe = Pipeline([
        ("scaler", StandardScaler()),
        ("mlp", MLPClassifier(
            hidden_layer_sizes=(64, 32),
            activation="relu",
            max_iter=500,
            random_state=RANDOM_SEED,
            early_stopping=True,
            validation_fraction=0.15
        ))
    ])

    mlp_plain_pipe.fit(X_qual_train, y_qual_train)
    plain_pred = mlp_plain_pipe.predict(X_qual_test)
    plain_f1 = f1_score(y_qual_test, plain_pred, average="macro")

    mlp_enriched_pipe.fit(X_qual_train_enriched, y_qual_train)
    enriched_pred = mlp_enriched_pipe.predict(X_qual_test_enriched)
    enriched_f1 = f1_score(y_qual_test, enriched_pred, average="macro")

    print("\nMLP quality predictor comparison:")
    print(f"  Plain Test Macro-F1     : {plain_f1:.4f}")
    print(f"  Enriched Test Macro-F1  : {enriched_f1:.4f}")

    if enriched_f1 > plain_f1:
        print("  -> Embedding similarity improved quality prediction")
    else:
        print("  -> Embedding similarity did not help significantly")

    print("\nClassification report (enriched model):")
    print(classification_report(
        y_qual_test,
        enriched_pred,
        target_names=["wrong", "correct"],
        zero_division=0
    ))

else:
    print("Skipping enriched quality predictor (embeddings or required variables not available)")
```

Adding the embedding similarity feature did not improve the quality predictor, confirming that the correctness of predictions is inherently difficult to forecast from input features alone.

## Summary of Findings

**Best model:** Fine-tuned BioBERT achieved the highest Test Macro-F1 of **0.4108** (50% accuracy), outperforming all traditional classifiers. However, McNemar's test (p=0.54) indicates this improvement over the baseline is **not statistically significant**, meaning the difference could be due to chance on this 200-sample test set.

**Key findings from each experiment:**

**Axis 1 — Text Representation:**
- Question-only (F1: 0.33) performed comparably to Question+Context (F1: 0.33), suggesting that for simple TF-IDF models, the context introduces as much noise as signal.
- Adding the long answer did not help (F1: 0.32) — the additional text dilutes discriminative features in a bag-of-words setting.
- BioBERT sentence embeddings (F1: ~0.37) offered a marginal improvement, but dense representations did not dramatically outperform sparse TF-IDF, indicating the bottleneck lies in the classification task itself rather than text representation.

**Axis 2 — Classifier Comparison:**
- Tuned LR (F1: 0.28), Calibrated SVM (F1: 0.28), and Naive Bayes (F1: 0.30) all performed similarly with word+char TF-IDF features.
- SMOTE oversampling (F1: 0.35) showed high CV performance (0.84) but poor generalisation, confirming that the minority class problem stems from genuine label ambiguity rather than insufficient training samples.
- The soft voting ensemble (F1: 0.28) performed **worse** than individual models, revealing that the classifiers make highly correlated errors rather than complementary ones. Poorly calibrated NB probabilities likely distorted the vote.
- Fine-tuned BioBERT (F1: 0.41) was the strongest model, benefiting from class-weighted loss and domain-specific pre-training.

**Axis 3 — Classification vs LLM-based:**
- Zero-shot BART-MNLI (F1: 0.06) failed dramatically, predicting 'maybe' for 95% of samples due to a fundamental mismatch between the NLI framing and PubMedQA's yes/no/maybe labels. This strongly favours task-specific supervised classification over zero-shot LLM approaches for this dataset.

**Error Analysis:**
- The 'maybe' class is misclassified 81.8% of the time, with most errors going to 'yes'. Hedging language appears across all classes, making it an unreliable discriminator.
- 24% of test samples (48/200) are misclassified by **all** models, representing genuinely ambiguous cases.
- The learning curve shows severe overfitting (train F1: 1.0, val F1: 0.37), suggesting more labelled data could help.

**Quality Predictor:**
- Random Forest, MLP, and Logistic Regression quality predictors achieved only ~0.50-0.54 Macro-F1, confirming that surface-level features (text length, hedging, negation) cannot reliably predict which questions will receive correct answers. Adding BioBERT embedding similarity did not improve prediction.
